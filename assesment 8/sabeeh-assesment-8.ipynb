```python
def results(marks):
    if marks >= 90:
        return "A"
    elif marks >= 89 and marks <= 80:
        return "B"
    elif marks >=79 and marks <=70:
        return "C"
    else: return "Fail"

res=results(40)
print(res)
        
```

```python
def isEven(num):
    if num%2 == 0:
        return True
    else: return False

isEven(7)
```

```python
def factorial(n):
    num=1
    for i in range (n):
        num=num*n

    return num

res=factorial(3)
print(res)
    
```
